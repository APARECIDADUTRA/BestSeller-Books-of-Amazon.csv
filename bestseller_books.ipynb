**# Função para corrigir os valores na coluna Depression State **


```python
def corrigir_coluna(valor):
    try:
        if '\t' in valor:
            return valor.replace('\t', '')
        elif '2\t' in valor:
            return valor.replace('2\t', 'No')
        elif '5\t' in valor:
            return valor.replace('5\t', 'No')
        else:
            return valor
    except:
        return valor
```
